```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example: Load historical price data for Bitcoin and S&P 500 index
bitcoin_data = pd.read_csv('BTC-GBP_Data.csv', index_col='Date', parse_dates=True)
atom_data = pd.read_csv('ATOM-GBP_Data.csv', index_col='Date', parse_dates=True)

# Calculate daily returnsACA
bitcoin_returns = bitcoin_data['Close'].pct_change()
atom_returns = atom_data['Close'].pct_change()

# Concatenate returns into a DataFrame
returns_df = pd.concat([bitcoin_returns, atom_returns], axis=1)
returns_df.columns = ['Bitcoin Returns', 'Atom Returns']
# Drop rows with NaN values
returns_df.dropna(inplace=True)

# Compute correlation matrix
correlation_matrix = returns_df.corr()

# Plot correlation heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Correlation Heatmap: Bitcoin vs. ATOM Returns')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load combined data of all 30 coins into a single DataFrame
crypto_data = pd.read_csv('combined_crypto_data.csv', index_col='Date', parse_dates=True)

# Calculate daily returns for each cryptocurrency
crypto_returns = crypto_data.pct_change()

# Drop rows with NaN values
crypto_returns.dropna(inplace=True)

# Compute correlation matrix
correlation_matrix = crypto_returns.corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Correlation Heatmap: Cryptocurrency Returns')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```
